```python
import requests
import json
from pprint import pprint
from nytkey import key
import time
import datetime as dt
import numpy as np
import pandas as pd
from grkeys import grkey
```

```python
#Return all bestseller lists 
nyturl1 = f"https://api.nytimes.com/svc/books/v3/lists/names.json?api-key={key}"

lists = requests.get(nyturl1).json()
pprint(lists)
```

```python
#Store a list of active bestseller lists
bslists = []

for x in range(55):
    if "2019" in lists['results'][x]['newest_published_date']:
        bslists.append(lists['results'][x]['display_name'])
        
print(bslists)
```

```python
#Store a list of dates to query for each list over the last 3 years:

dates = []

current_date = "2019-04-28"
current_date = dt.datetime.strptime(current_date, "%Y-%m-%d")
current_date = dt.datetime.date(current_date)

for x in range(156):
    pastdate = current_date - (dt.timedelta(days=7))*x
    dates.append(pastdate)

#Transform all datetime objects back into strings so we can retrieve:
string_dates = []

for d in dates:
    temp = d.strftime("%Y-%m-%d")
    string_dates.append(temp)

  
```

```python
#Retrieve 'Combined Print & Ebook Fiction'
combined = []

for string in string_dates:
    nytcombinedurl = f"https://api.nytimes.com/svc/books/v3/lists/{d}/combined-print-and-e-book-fiction.json?api-key={key}"
    data = requests.get(nytcombinedurl).json()
    combined.append(data)
    time.sleep(6)
```

```python
pprint(combined)
```

```python
#Convert the json results into a dataframe and save for future analysis
titles = []
authors = []
publishers = []
isbn10 = []
isbn13 = []
descriptions = []
book_images = []
#dates = []
for x in range(156):
    for y in range(10):
        try:
            titles.append(combined[x]['results']['books'][y]['title'])
            authors.append(combined[x]['results']['books'][y]['author'])
            publishers.append(combined[x]['results']['books'][y]['publisher'])
            isbn10.append(combined[x]['results']['books'][y]['primary_isbn10'])
            isbn13.append(combined[x]['results']['books'][y]['primary_isbn13'])
            descriptions.append(combined[x]['results']['books'][y]['description'])
            book_images.append(combined[x]['results']['books'][y]['book_image'])
            #dates.append(combined[x]['results']['bestsellers_date'])
        except KeyError:
            print("Data Not Found")
            
```

```python
nyt_df = pd.DataFrame({'Title' : titles, 'Author': authors, 'Publisher' : publishers, 'ISBN-10' : isbn10, 'ISBN-13' : isbn13, "Description" : descriptions, 'Image' : book_images})

nyt_df.to_csv("nytcombinedbestsellers.csv", index=False, header=True)
```
